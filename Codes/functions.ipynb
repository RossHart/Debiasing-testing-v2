```python
def f_logistic(x, k, c):
    # Function to fit the data bin output from the raw plot function
    L = (1 + np.exp(c))
    r = L / (1.0 + np.exp(-k * x + c))
    return r


def f_exp_pow(x, k, c):
    # Function to fit the data bin output from the raw plot function
    r = np.exp(-k * (-x) ** c)
    return r
```

```python
logistic_bounds = ((0.5, 10), (-10, 10))
exponential_bounds = ((10**(-5),10),(10**(-5),10))

xg = np.linspace(-2,0,100)
```

```python
for k in [0.2,10]:
    for c in [-1,1]:
        plt.plot(xg,f_logistic(xg,k,c),label='k={},c={}'.format(k,c))

plt.legend()
```
